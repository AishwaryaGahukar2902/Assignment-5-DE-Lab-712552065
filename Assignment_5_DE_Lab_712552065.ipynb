```python
import pandas as pd

data = {
'user_id':[101,102,103,104,105,106,107,108,109,110,111,112,113,114,115,116,117,118,119,120],
'device_type':["Android","iOS","Android","iOS","Web","Android","iOS","Web","Android","iOS",
               "Android","Web","iOS","Android","Web","Android","iOS","Android","Web","iOS"],
'country':["India","USA","Brazil","India","UK","India","USA","Canada","Germany","India",
           "USA","Brazil","UK","India","USA","Canada","Germany","India","UK","USA"],
'followers':[1200,5000,800,15000,2000,900,3200,1100,2400,18000,950,700,4200,1300,2600,1000,3500,850,3000,22000],
'likes':[150,420,95,1200,260,80,350,140,300,2000,120,65,500,160,310,110,420,90,330,2500],
'comments':[12,35,8,110,20,5,28,10,22,150,9,5,40,12,22,7,35,6,25,180],
'share_count':[5,12,3,30,9,2,14,4,11,40,4,2,20,5,8,3,12,3,10,50],
'watch_time':[34,58,22,150,46,18,60,30,62,200,28,15,75,32,58,26,78,20,65,240],
'video_length':[60,90,40,180,70,35,85,60,100,220,50,30,95,55,80,45,100,38,90,260]
}

df = pd.DataFrame(data)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

# 1 . HISTOGRAM â€“ Distribution of Likes

```python
plt.figure(figsize=(8,5))
plt.hist(df['likes'], bins=15)
plt.title("Histogram of Likes")
plt.xlabel("Likes")
plt.ylabel("Frequency")
plt.show()
```

# 2. BAR CHART â€“ Average Watch Time by Device Type

```python
plt.figure(figsize=(8,5))
df.groupby('device_type')['watch_time'].mean().plot(kind='bar')
plt.title("Average Watch Time by Device Type")
plt.xlabel("Device Type")
plt.ylabel("Watch Time (seconds)")
plt.show()
```

# 3. SCATTER PLOT â€“ Likes vs Comments

```python
plt.figure(figsize=(8,5))
plt.scatter(df['likes'], df['comments'])
plt.title("Scatter Plot: Likes vs Comments")
plt.xlabel("Likes")
plt.ylabel("Comments")
plt.show()
```

# 4. PIE CHART â€“ Country Distribution

```python
plt.figure(figsize=(7,7))
df['country'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("User Distribution by Country")
plt.ylabel("")
plt.show()
```

# 5. HEATMAP â€“ Correlation between Numerical Features

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# 6. PAIR PLOT â€“ Likes, Comments, Shares, Watch Time

```python
sns.pairplot(df[['likes','comments','share_count','watch_time']])
plt.show()
```

# 7. VIOLIN PLOT â€“ Watch Time by Device Type

```python
plt.figure(figsize=(8,5))
sns.violinplot(x=df['device_type'], y=df['watch_time'])
plt.title("Violin Plot: Watch Time vs Device Type")
plt.xlabel("Device Type")
plt.ylabel("Watch Time (seconds)")
plt.show()
```
